# Eksik Veri Komutları 
1. isnull() veya isna(): Eksik değerler için True döndürür.
2. notnull() veya notna(): isnull'un(veya isna'nın) tam tersi.
3. dropna(): Eksik verileri kaldırır.
4. fillna(): Eksik verileri doldurur.

```python
data = pd.Series([1,2,np.nan,'hello',None,61])
data
```

```python
data.isnull()
```

```python
data.notnull()
```

```python
data.isna()
```

```python
data[data.notna()] # Eksik olmayan verileri alır.
```

```python
Q1=df.Salary.quantile(0.25) # çeyreklere ayırmaya yarayan fonskiyon, Q1 = 1.çeyrek, yani dikdörtgenin sağ aşağı köşe değeri
Q3=df.Salary.quantile(0.75) # Q3= 3.çeyrek, yani dikdörtgenin sağ yukarı köşe değeri
IQR=Q3-Q1 # medyan değerler.(yeşil çizgi)
Q1, Q3, IQR
```

```python
Q3 + 1.5*IQR # Bu grafik için bu değerden büyük değerlere aykırı veri diyeceğiz.
```

# Z skoruna göre aykırı veri analizi yaklaşımı
Verinin ortalamadan kaç standart sapma saptığını göstermeye yarar

```python
from scipy import stats # İstatistik modülü
```

```python
z=np.abs(stats.zscore(df.Salary)) #stats.zscore salary içersindeki değerlerin ortalamadan kaç standart sapma saptığını gösterir.
```

```python
z
```

```python
np.where(z>2)[0].shape # np.where(), mantıksal sınama yapar ve True olduğu durumları getirir.
# Tuple içerisinde array bulunur bu yüzden işlem yapmak için array'i almamız gerek. Bu yüzde[0] demeliyiz.
# Aykırı veri 1 tane bulundu
```

```python
df[~(z>2)] # ~ 'değildir' işareti ile Aykırı olmayan verileri listemeiş olduk
```

```python
df[(z>2)] # Aykırı verimiz.
```
